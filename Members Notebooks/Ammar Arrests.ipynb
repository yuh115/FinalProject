### Arrests frequencey around neighborhoods in Allegheny county
An arrest is the act of taking a person into custody by law enforcement officers for the purpose of charging them with a crime. The arrest is made when a police officer has probable cause to believe that the person has committed a crime. A higher number of arrests in a neighborhood can have an impact on the safety of the area. If a neighborhood has a high number of arrests, it may suggest that there is a high level of criminal activity in that area, which can make residents feel unsafe. Additionally, a high number of arrests can put a strain on the resources of law enforcement agencies, making it harder for them to respond to emergencies and prevent future crimes. Overall, a high number of arrests in a neighborhood can have negative consequences on the safety and wellbeing of the community.


```python
import geopandas as gpd
from shapely.geometry import Point
import pandas as pd

neighborhoods = gpd.read_file("https://data.wprdc.org/dataset/e672f13d-71c4-4a66-8f38-710e75ed80a4/resource/c5a93a8e-03d7-4eb3-91a8-c6b7db0fa261/download/pit")
neighborhoods.head(3)
#pd.DataFrame(neighborhoods["hood"])
```

```python
arrests = pd.read_csv("https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f")
arrests.head(3)
```

```python

dic = {}

# iterating through the elements of list
for i in neighborhoods["hood"]:
    dic[i] = 0
    
for index, row in arrests.iterrows():
    arrests_neighborhood =  row["INCIDENTNEIGHBORHOOD"]
    if arrests_neighborhood in dic:
        dic[arrests_neighborhood] += 1
        
for n in dic:
    print(n + ": " + str(dic[n]))
```

```python
df = pd.DataFrame.from_dict(dic, orient="index", columns = ["Arrests"])
display(df)
```

```python
df = df.sort_values("Arrests", ascending=True)
df.iloc[:25].plot(kind = "bar", figsize = (10, 10))
```

```python
df.iloc[:20].plot.pie(y='Arrests', figsize=(15, 15))
```
